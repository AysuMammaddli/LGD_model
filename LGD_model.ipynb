## 7. Korrelasiya cedveli qurulur (data distribution-a uyqun olaraq)

#### Thresholds 

1. target uzre 10%
2. intercorrelation uzre 60%

```python
def target_correlation(data, target, threshold=0.1):
    corr_data = data.corr(numeric_only=True)
    
    target_corr=corr_data[target].abs()>threshold
    
    variables_explaining_target = target_corr[target_corr].index.tolist()

    variables_explaining_target.remove(target)
    
    result_explaining_target = pd.DataFrame(
    {'Variable': variables_explaining_target,
    'Correlation with Target': corr_data.loc[variables_explaining_target, target]})
    
    result_explaining_target.reset_index(drop=True, inplace=True)
    
    return result_explaining_target
    
```

```python
r=target_correlation(data, 'LGD ($)')
r
```

```python
def intercorrelation(data, target, threshold=0.6):
    corr_data=data.drop(columns=[target]).corr(numeric_only=True)
    highly_corr_variables=(np.abs(corr_data)>threshold)&(np.abs(corr_data)!=1)
    independent_var=np.where(highly_corr_variables)
    
    result=pd.DataFrame({'variable1':corr_data.index[independent_var[0]],
                        'variable2':corr_data.index[independent_var[1]]})
    return result
```

```python
intercorrelation(data, 'LGD ($)')
```

## 8. VIF statistik deyerlerine baxilir

Yalniz musteqil numeric deyishenlerin oz aralarinda asilliq yoxlanilir. 1-9 intervalinda olanlar saxlanilir.

1-9 intervalindan yuksek netice veren numeric musteqil deyishenler data-dan drop olunur.

```python
from statsmodels.stats.outliers_influence import variance_inflation_factor
def vif(data, variables):
    vif_data=data[variables]
    vif_values=[variance_inflation_factor(vif_data.values, i) for i in range(vif_data.shape[1])]
    vif_result=pd.DataFrame({'VIF':vif_values, 'features':variables})
    return vif_result

vif_result=vif(data, ['Exposure Amount ($)',                               
#'Credit Score'  ,                 
'Loan Term (Months)' ,               
#'Interest Rate (%)'  ,                
'Income ($)'          ,           
#'Loan to Value Ratio (%)',          
'Debt to Income Ratio (%)',          
'Employment History (Years)',
'Previous Defaults'])

vif_result
```

## 16. Univariate analysis

her bir musteqil deyishen ile asili deyishen arasinda xetti reqressiya modeli qurularaq ferdi R2 accuracylerine baxilir

musteqil deyishenler individual olaraq modele tesiri yoxlanilir

test uzre r2 neticesi 1%den (0.01) yuxari olan ve overfit etmeyen musteqil deyishenler sechilir

```python
result=[]
for i in x_train.columns:
    x_train_single=x_train[[i]]
    x_test_single=x_test[[i]]
    reg.fit(x_train_single, y_train)
    y_pred_train_single=reg.predict(x_train_single)
    train_r2=metrics.r2_score(y_train, y_pred_train_single)
    y_pred_test_single=reg.predict(x_test_single)
    test_r2=metrics.r2_score(y_test, y_pred_test_single)
    
    result.append({'Variable': i, 'Train R2': train_r2, 'Test R2': test_r2})
result_df = pd.DataFrame(result)
result_df_sorted = result_df.sort_values(by='Test R2', ascending=False)
result_df_sorted
```
